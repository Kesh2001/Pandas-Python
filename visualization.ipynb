# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('D:\Study Material\Data Analyst\Python\Pandas\Ice Cream Ratings.csv')
df
df.set_index('Date')
```

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8')
```

```python
df.plot(kind='line', title='Ice Cream Ratings', xlabel='Daily Rating', ylabel= 'Scores')
```

```python
df.plot.barh(stacked=True)
```

```python
df.plot.scatter(x='Texture Rating',y='Overall Rating', s=100, c='Yellow') # s is size
```

```python
df.plot.hist(bins=100)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize=(10,5))
```

```python
df.plot.pie(y='Flavor Rating', figsize=(10,6))
```
